```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'Asst.csv'# Load the CSV file
df = pd.read_csv(file_path)
```

```python
df
```

```python
df_cleaned = df.iloc[:, 0].str.split(',', expand=True)# Split the data into individual columns based on commas
df_cleaned.columns = ['Date', 'Share Name', 'Open', 'High', 'Low', 'Close', 'Volume']# Assign proper column names
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], errors='coerce')# Convert the 'Date' column to datetime format
df_cleaned['Close'] = pd.to_numeric(df_cleaned['Close'], errors='coerce')# Convert 'Close' column to numeric
df_cleaned = df_cleaned.dropna(subset=['Date', 'Close']) # Drop rows with NaN values in the 'Date' or 'Close' columns
```

```python
df_cleaned
```

# a

```python
# Plot the closing prices over time
plt.figure(figsize=(10, 6))
plt.plot(df_cleaned['Date'], df_cleaned['Close'], label='Closing Price', color='blue')
# plt.xlabel('Date')
# plt.ylabel('Closing Price')
# plt.title('Closing Prices Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

# b

```python
# Customize plot style to resemble graph paper
plt.figure(figsize=(12, 6))
plt.plot(df_cleaned['Date'], df_cleaned['Close'], label='Closing Price', color='blue', marker='o', linestyle='-', linewidth=2, markersize=5)
# Add labels and title
plt.xlabel('Date', fontsize=14)
plt.ylabel('Closing Price (USD)', fontsize=14)
plt.title('XYZ Corp. Closing Prices Over Time', fontsize=16, fontweight='bold')
# Customize grid to look like graph paper
plt.grid(True, which='both', color='gray', linestyle='-', linewidth=0.5)
plt.minorticks_on()  # Enable minor ticks (for a finer grid)
plt.grid(which='minor', color='lightgray', linestyle='--', linewidth=0.5)
# Add x and y ticks for more of a graph paper feel
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
# Add a legend
plt.legend(fontsize=12)
# Ensure tight layout
plt.tight_layout()
# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'Asst.csv'
df = pd.read_csv(file_path)

# Split the data into individual columns based on commas
df_cleaned = df.iloc[:, 0].str.split(',', expand=True)

# Assign proper column names
df_cleaned.columns = ['Date', 'Share Name', 'Open', 'High', 'Low', 'Close', 'Volume']

# Convert relevant columns to numeric
df_cleaned['Open'] = pd.to_numeric(df_cleaned['Open'], errors='coerce')
df_cleaned['High'] = pd.to_numeric(df_cleaned['High'], errors='coerce')
df_cleaned['Low'] = pd.to_numeric(df_cleaned['Low'], errors='coerce')
df_cleaned['Close'] = pd.to_numeric(df_cleaned['Close'], errors='coerce')

# Drop rows with NaN values
df_cleaned = df_cleaned.dropna(subset=['Open', 'High', 'Low', 'Close'])

# Prepare data in long form for Seaborn box plot
df_melted = df_cleaned.melt(value_vars=['Open', 'High', 'Low', 'Close'], 
                            var_name='Price Type', value_name='Price')
```

# a

```python
sns.boxplot(x='Price Type', y='Price', data=df_melted)
plt.show()
```

# b

```python
sns.boxplot(x='Price Type', y='Price', data=df_melted, palette='Set2')


# Add labels and title
plt.title('Box Plot of Stock Prices', fontsize=20)
plt.xlabel('Price Type', fontsize=14)
plt.ylabel('Price (BDT)', fontsize=14)
# Improve appearance
plt.grid(True, linestyle='-', linewidth=0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
