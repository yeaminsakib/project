```python
import geopandas as gpd

# Load world geometries from Natural Earth via Geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Select a country (replace 'Bangladesh' with your desired country name)
country_name = "Bangladesh"
country = world[world['name'] == country_name]

# Example population and area data for the country (replace with actual data)
population_data = {
    'Bangladesh': 171200000  # Replace with actual population for your country
}

area_data = {
    'Bangladesh': 147570  # Replace with actual area (in square kilometers)
}

# Add population and area columns
country['population'] = country['name'].map(population_data)
country['area_km2'] = country['name'].map(area_data)

# Calculate population density (people per square kilometer)
country['pop_density'] = country['population'] / country['area_km2']

# Save the updated country's geometry and population density to a shapefile
country.to_file(f"{country_name}_shapefile_with_pop_density.shp")

# Output the GeoDataFrame to verify the 'pop_density' column
print(country[['name', 'population', 'area_km2', 'pop_density']])
```
